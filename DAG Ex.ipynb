```python
import numpy as np
import pandas as pd
```

```python
def extract_csv(file_path):
    try:
        #columns to extract
        extract_cols = ['state', 'county', 'site name', 'naics description', 'site latitude',
           'site longitude', 'city', 'pollutant desc', 'pollutant type(s)',
           'total emissions']
        return pd.read_csv(file_path, usecols=extract_cols)
    except:
        print("Something wrong!!!!! Check file path.")
```

```python
df = extract_csv('point_12345.csv')
```

```python
df2 = extract_csv('point_678910.csv')
```

```python
sdf = pd.concat([df, df2], ignore_index=True)
```

```python
from sqlalchemy import create_engine
# PostgreSQL database connection
postgresql_engine = create_engine('postgresql://postgres:1352@localhost:5432/epa')
```

```python
fdf = sdf.groupby(['state','county','city'], as_index=False).sum()['total emissions']
```

```python
fdf.to_sql('geo_poll_lvl', con=postgresql_engine,index=False)
```

```python
fdf= sdf.groupby(['pollutant desc').sum()[['total emissions']]
fdf.to_sql('pollutant', con=postgresql_engine,index=False)
```
